```python
#Save every columns but the name into a separate dataframe
moviestats = movies[["IMDB Rating","Biography","Drama","Thriller","Comedy","Crime","Mystery","History"]]
```

```python
moviestats
```

```python
#turn the movie stats above into a matrix which will later be used for the knn algorithm
mat_movie_users = csr_matrix(moviestats.values)
```

```python
#Not many dimensions and rows, that is why I chose the brute force method
#Set N neighbors to 5 since we are recommending 5 movies
#Utilizing the euclidean distance formula
model_knn = NearestNeighbors(metric='euclidean', algorithm='brute', n_neighbors=5)
```

```python
#utilize the nearest neighbor algorithm on the stats of all the movies, excluding The Post
model_knn.fit(mat_movie_users)
```

```python
newmoviestats = [7.2,1,1,0,0,0,0,1]
movienameid = [24, "The Post"]
```

```python
def movierecommendation(movienums, nameid):
  x = np.array([7.2,1,1,0,0,0,0,1])
  x = x.reshape(1,-1)
  distances, indices=model_knn.kneighbors(x)
  for i in indices:
      print(movies['Movie Name'][i])
  #for appending purposes
  moviestats.loc[len(moviestats)] = movienums
  #fullmovielist = nameid + movienums
  movies.loc[len(movies)] = nameid + movienums
```

```python
movierecommendation(newmoviestats, movienameid)
```
